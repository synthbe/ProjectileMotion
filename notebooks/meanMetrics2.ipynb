```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
sphere = pd.read_csv('../tables/sphere1_2.csv', sep=',')
sphere
```

```python
sphere.columns
```

```python
tidy_sphere = sphere.stack().reset_index()
tidy_sphere.drop('level_0', axis=1, inplace=True)
tidy_sphere.columns = ["h", "range"]
tidy_sphere.head(10)
```

```python
sns.displot(data=tidy_sphere, col="h", kind='hist', x='range')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(15, 5))
for i, col in enumerate(sphere.columns):
    sns.histplot(data=sphere[col], bins=20, kde=True, ax=axes[i]) 

plt.tight_layout()
plt.show()
```
