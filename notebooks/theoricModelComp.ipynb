```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sphere1 = pd.read_csv("../tables/sphereDay1_1.csv")
sphere2 = pd.read_csv("../tables/sphereDay1_2.csv")
sphere3 = pd.read_csv("../tables/sphereDay2_1.csv")
```

```python
sphere1.shape, sphere2.shape, sphere3.shape
```

```python
spheres = [sphere1, sphere2, sphere3]
launchHeights = ['30.00', ' 40.00']
planeHeights = [17.80, 18.80]
```

```python
def calculateRange(launchHeight: float, planeHeight: float, mass: float) -> float:
  # Provavelmente esta errado!!!
  g = 9.80665
  H = launchHeight / 100
  h = planeHeight / 100
  deltaH = H - h
  m = mass / 1000
  t = np.sqrt(2 * h / g)
  
  r = m * g * deltaH * t * 100

  return r

calculateRange(30, 17.80, 22)
```

```python
sns.lineplot()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 10))
for i in range(3):
  for j in range(2):
    ax = axes[i, j]
    ax.set_title(f"Esfera {i+1}")
    currSphere = spheres[i]
    ranges = currSphere[launchHeights[j]]
    sns.scatterplot(ranges, ax=ax)
    sns.lineplot(np.full((10,), np.mean(ranges)), ax=ax, label="Valor Teórico")

plt.tight_layout()
plt.legend()
plt.show()
```
